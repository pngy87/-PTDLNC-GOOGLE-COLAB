```python
#load data
import gdown
url = "https://drive.google.com/uc?id=14kGMTPKZKjFJEDZj3DCMCjXR2mEUJrhP"
data_filename = "Sample - Superstore.xls"
gdown.download(url, data_filename, quiet=False)
```

```python
#Read data
import pandas as pd
df = pd.read_excel(data_filename)
df
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(10, 6))
df["Sales"].plot(kind="hist", bins=10000)
plt.xscale('log')
plt.title("Sales Distribution (Log Scale)")
plt.xlabel("Sales (Log Scale)")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

```python
# give code density plot of sale (tần suất)
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
sns.kdeplot(df['Sales'], fill=True)
plt.title('Density Plot of Sales')
plt.xlabel('Sales')
plt.ylabel('Density')
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.regplot(x="Sales", y="Profit", data=df, scatter_kws={'alpha':0.6, 's':10}, line_kws={'color':'red'})
plt.title("Sales vs Profit with Trend Line")
plt.xlabel("Sales")
plt.ylabel("Profit")
plt.tight_layout()
plt.show()
```

# Visualization

```python
import matplotlib.pyplot as plt
```

```python
df.set_index("Order Date")["Profit"].resample("W").sum()
```

```python
# 1) Line (Series.plot): Monthly Sal
df.set_index("Order Date")["Sales"].resample("ME").sum().to_frame("Monthly_Sales")
```

```python
# 1) Line (Series.plot): Monthly Sales
monthly_sales = df.set_index("Order Date")["Sales"].resample("ME").sum()
monthly_sales.index = monthly_sales.index.to_period("M").astype(str)
print(monthly_sales.head())
monthly_sales.plot(grid= True, figsize=(10, 5))
#
## help Plot(): https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.plot.html
```

```python
# 1) Line (Series.plot): Monthly Sales
monthly_sales = df.set_index("Order Date")["Sales"].resample("ME").sum()
monthly_sales.plot()
```

```python
# 2) Bar: Sales by Category
plt.figure(); df.groupby("Category")["Sales"].sum().sort_values(ascending=False).plot(kind="bar")
plt.title("Sales by Category"); plt.xlabel("Category"); plt.ylabel("Sales"); plt.tight_layout(); plt.show()
```

```python
# 2) Bar: Sales by Category
df.groupby("Category")["Sales"].sum().plot(kind="bar")
```

```python
# 3) H-Bar: Top 10 Sub-Category by Sales
import matplotlib.pyplot as plt
plt.figure(); df.groupby("Sub-Category")["Sales"].sum().nlargest(10).sort_values().plot(kind="barh")
plt.title("Top 10 Sub-Category by Sales"); plt.xlabel("Sales"); plt.ylabel("Sub-Category"); plt.tight_layout(); plt.show()
```

```python
# 4) Stacked bar: Region × Segment (Sales)
pvt = df.pivot_table(index="Region", columns="Segment", values="Sales", aggfunc="sum", fill_value=0)
plt.figure(); pvt.plot(kind="bar", stacked=True); plt.title("Sales by Region × Segment"); plt.xlabel("Region"); plt.ylabel("Sales"); plt.tight_layout(); plt.show()
```

```python
# 5) Histogram: Sales distribution
plt.figure();
df["Sales"].plot(kind="hist", bins=50);
plt.title("Sales Distribution"); plt.xlabel("Sales"); plt.ylabel("Frequency"); plt.tight_layout(); plt.show()
```

```python
# 6) Boxplot: Profit by Category
plt.figure(); df.boxplot(column="Profit", by="Category"); plt.title("Profit by Category"); plt.suptitle(""); plt.xlabel("Category"); plt.ylabel("Profit"); plt.tight_layout(); plt.show()
```

```python
# 7) Scatter: Sales vs Profit
plt.figure(); df.plot(kind="scatter", x="Sales", y="Profit", s=10, alpha=0.6)
plt.title("Sales vs Profit"); plt.xlabel("Sales"); plt.ylabel("Profit"); plt.tight_layout(); plt.show()
```

```python
# 8) Pie: Sales share by Region
plt.figure(); df.groupby("Region")["Sales"].sum().plot(kind="pie", autopct="%1.1f%%")
plt.title("Sales Share by Region"); plt.ylabel(""); plt.tight_layout(); plt.show()
```

```python
# 9) Area: Monthly Sales by Category (12 tháng cuối)
monthly_cat = (df.set_index("Order Date")
                 .groupby([pd.Grouper(freq="ME"), "Category"])["Sales"].sum()
                 .unstack(fill_value=0))
monthly_cat.index = monthly_cat.index.to_period("M").astype(str)
plt.figure(); monthly_cat.tail(12).plot(kind="area", stacked=True)
plt.title("Monthly Sales by Category (Last 12 Months)"); plt.xlabel("Month"); plt.ylabel("Sales"); plt.xticks(rotation=45); plt.tight_layout(); plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

#plot heatmap
# --- 1. Correlation Heatmap (cho biến numerical) ---
plt.figure(figsize=(6,4))
corr = df[["Sales", "Quantity", "Discount", "Profit"]].corr()
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap (Sales, Quantity, Discount, Profit)")
plt.show()

# --- 2. Crosstab Heatmap (Region × Category, đếm số đơn hàng) ---
ct = pd.crosstab(df["Region"], df["Category"])
plt.figure(figsize=(6,4))
sns.heatmap(ct, annot=True, cmap="YlGnBu", fmt="d")
plt.title("Crosstab Heatmap (Region × Category)")
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
df.groupby("State")["Sales"].sum().sort_values(ascending=False).plot(kind="bar")
plt.title("Total Sales by State")
plt.xlabel("State")
plt.ylabel("Total Sales")
plt.tight_layout()
plt.show()
```
