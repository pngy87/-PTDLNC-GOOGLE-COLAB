##1.2 missing data

```python
# Function to analyse the missing data

def analyze_null_values(df):
    missing_data = df.isnull().sum().to_frame(name="Null Values")
    missing_data["Total Values"] = len(df)
    missing_data["Percentage of Missing Values"] = (missing_data["Null Values"] / missing_data["Total Values"]) * 100
    return missing_data.sort_values(by="Percentage of Missing Values", ascending=False)

# Example usage
analyze_null_values(tx)
```

```python
import pandas as pd
import numpy as np


#  - loại bổ các cột ID, mã số)
num_cols = [c for c in tx.select_dtypes(include=[np.number]).columns
            if c not in {"transaction_id", "product_id", "customer_id"}]

# 2) Hàm tính ngưỡng IQR
def iqr_bounds(s: pd.Series, k: float = 1.5):
    q1 = s.quantile(0.25)
    q3 = s.quantile(0.75)
    iqr = q3 - q1
    lower = q1 - k * iqr
    upper = q3 + k * iqr
    return lower, upper, iqr

# 3) Báo cáo outliers theo từng cột
outlier_report = []
for col in num_cols:
    x = tx[col].dropna()
    if x.empty:
        continue
    lower, upper, iqr = iqr_bounds(x, k=1.5)
    mask = (tx[col] < lower) | (tx[col] > upper)
    outlier_report.append({
        "column": col,
        "n": int(tx[col].notna().sum()),
        "q1": x.quantile(0.25),
        "q3": x.quantile(0.75),
        "iqr": iqr,
        "lower_bound": lower,
        "upper_bound": upper,
        "n_outliers": int(mask.sum()),
        "outlier_ratio_%": 100 * mask.mean()
    })

outlier_df = pd.DataFrame(outlier_report).sort_values("n_outliers", ascending=False)
print("=== Outlier summary (IQR=1.5) ===")
print(outlier_df.to_string(index=False))

# 4) Xử lý outliers: 2 lựa chọn phổ biến

# 4A) CAPPING (Winsorization) — chặn giá trị vượt ngưỡng về sát ngưỡng
tx_capped = tx.copy()
for col in num_cols:
    x = tx[col]
    if x.notna().sum() == 0:
        continue
    lower, upper, _ = iqr_bounds(x.dropna(), k=1.5)
    tx_capped[col] = x.clip(lower=lower, upper=upper)

# 4B) REMOVE (Drop rows) — bỏ dòng có outlier (cẩn thận vì có thể mất dữ liệu nhiều)
tx_dropped = tx.copy()
for col in num_cols:
    x = tx_dropped[col]
    if x.notna().sum() == 0:
        continue
    lower, upper, _ = iqr_bounds(x.dropna(), k=1.5)
    tx_dropped = tx_dropped[(tx_dropped[col].isna()) | ((tx_dropped[col] >= lower) & (tx_dropped[col] <= upper))]

print("Original shape:", tx.shape)
print("Capped shape  :", tx_capped.shape)
print("Dropped shape :", tx_dropped.shape)
```

## 3.1.4	Fixing Errors (Sửa lỗi dữ liệu)

1. Chuỗi: loại ký tự không in được, strip() khoảng trắng thừa.

2. Kiểu dữ liệu: ép *date* → datetime64; các cột giá/chi phí/số lượng → numeric.

3. online_order: chuẩn hóa về boolean (True/False) nếu có thể suy ra.

4. order_status: chuẩn hóa về tập nhãn hợp lệ (Approved, Cancelled, Completed, Pending, Disputed; còn lại giữ Title dạng chuẩn).

5. Số âm không hợp lệ trong list_price/standard_cost/unit_price/amount → đặt NaN để xử lý tiếp.

6. Ngày logic: nếu product_first_sold_date > transaction_date thì sửa = transaction_date.

7. transaction_date tương lai → đặt NaT (để xem xét lại).

8. Gắn cờ (không tự sửa) các dòng có list_price < standard_cost (để review thủ công).

```python
# @title Default title text
# 1) Chuẩn hoá tên cột:
#    - strip: bỏ khoảng trắng thừa
#    - replace: thay khoảng trắng bằng "_"
#    - lower: về chữ thường
tx.columns = (tx.columns.astype(str)
              .str.strip()
              .str.replace(r"\s+", "_", regex=True)
              .str.lower())

# Sao lưu dữ liệu gốc để so sánh trước/sau
tx_before = tx.copy()

# 2) Xử lý các cột chuỗi (object)
#    - convert sang str
#    - loại bỏ ký tự không in được (regex)
#    - strip: bỏ khoảng trắng đầu/cuối
for col in tx.select_dtypes(include="object").columns:
    tx[col] = (tx[col].astype(str)
                        .str.replace(r"[\x00-\x1f\x7f]", "", regex=True)
                        .str.strip())

# 3) Parse các cột có chữ 'date' thành datetime
for col in [c for c in tx.columns if "date" in c]:
    tx[col] = pd.to_datetime(tx[col], errors="coerce")

# 4) Chuyển các cột numeric (có từ khóa price/cost/amount/qty/…) sang số
num_like = [c for c in tx.columns
            if any(k in c for k in ["price","cost","amount","qty","quantity","profit","revenue","unit"])]
for col in num_like:
    tx[col] = pd.to_numeric(tx[col], errors="coerce")

# 5) Khởi tạo log để lưu lại các bước sửa lỗi
fix_log = []

# 5.1 online_order: chuẩn hóa về boolean (True/False)
if "online_order" in tx.columns:
    before_na = tx["online_order"].isna().sum()
    tx["online_order"] = (tx["online_order"]
                          .map({True: True, False: False, "True": True, "False": False,
                                "TRUE": True, "FALSE": False})
                          .astype("boolean"))
    fix_log.append({"rule": "Normalize online_order to boolean",
                    "affected_rows": int(before_na - tx["online_order"].isna().sum())})

# 5.2 order_status: chuẩn hóa về tập hợp hợp lệ (Approved, Cancelled, …)
if "order_status" in tx.columns:
    valid_status = {
        "Approved": {"approved", "approve", "ok", "accepted"},
        "Cancelled": {"cancelled", "canceled", "void"},
        "Completed": {"completed", "complete", "done"},
        "Pending": {"pending", "in progress", "processing", "awaiting"},
        "Disputed": {"disputed", "dispute"}
    }
    def normalize_status(x):
        if pd.isna(x):
            return np.nan
        s = str(x).strip().lower()
        for k, vals in valid_status.items():
            if s in vals or s == k.lower():
                return k
        return s.title() if s else np.nan  # giữ lại dạng Title nếu không thuộc nhóm nào

    before_vals = tx["order_status"].copy()
    tx["order_status"] = tx["order_status"].apply(normalize_status)
    changed = (before_vals != tx["order_status"]).sum()
    fix_log.append({"rule": "Normalize order_status labels", "affected_rows": int(changed)})

# 5.3 Các giá trị số âm bất hợp lý (list_price, cost, …) → đặt NaN
for col in [c for c in ["list_price","standard_cost","unit_price","amount"] if c in tx.columns]:
    neg_mask = tx[col] < 0
    n_neg = int(neg_mask.sum())
    if n_neg > 0:
        tx.loc[neg_mask, col] = np.nan
        fix_log.append({"rule": f"Set negatives to NaN in {col}", "affected_rows": n_neg})

# 5.4 Nếu product_first_sold_date > transaction_date → sửa = transaction_date
if {"product_first_sold_date","transaction_date"}.issubset(tx.columns):
    bad_mask = (tx["product_first_sold_date"].notna() &
                tx["transaction_date"].notna() &
                (tx["product_first_sold_date"] > tx["transaction_date"]))
    n_bad = int(bad_mask.sum())
    if n_bad > 0:
        tx.loc[bad_mask, "product_first_sold_date"] = tx.loc[bad_mask, "transaction_date"]
        fix_log.append({"rule": "product_first_sold_date > transaction_date -> set to transaction_date",
                        "affected_rows": n_bad})

# 5.6 (Flag) Nếu list_price < standard_cost → gắn cờ để review thủ công
flag_count = 0
if {"list_price","standard_cost"}.issubset(tx.columns):
    flag_mask = (tx["list_price"].notna() &
                 tx["standard_cost"].notna() &
                 (tx["list_price"] < tx["standard_cost"]))
    flag_count = int(flag_mask.sum())
    fix_log.append({"rule": "Flag rows where list_price < standard_cost (manual review)",
                    "affected_rows": flag_count})

# 5) Xuất báo cáo các bước sửa lỗi
fix_log_df = pd.DataFrame(fix_log)

print(fix_log_df)

```
